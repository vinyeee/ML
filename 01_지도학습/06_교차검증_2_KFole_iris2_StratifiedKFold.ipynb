### 1. 데이터 수집

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
```

### 데이터를 쉽게 이해하기 위해 df로 변경한다.

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=iris.target
df['species_name']=iris.target_names[iris.target]
df
```

### 1. KFold 교차검증

```python
# n_splits ==> 총 데이터  150개를 5개의 fold로 구분, 각 폴드마다 30개씩 저장된다. 
# kfold=KFold(n_splits=5,shuffle=False,random_state=None) 기본
# shuffle = False ==> data 섞지 않음
# random_state ==> seed 고정

kfold=KFold(n_splits=3,shuffle=False,random_state=None) 
```

```python
from sklearn.metrics import accuracy_score
count=0
for train_index, valid_index in  kfold.split(iris_data):# index 반환
    count+= 1
    label_train = df['species'].iloc[train_index]
    label_valid = df['species'].iloc[valid_index]
    #print(label_train, label_valid)
    
    print('교차검증 순서:{}'.format(count))
    print('학습 데이터 분포:',label_train.value_counts())
    print('검증 데이터 분포:',label_valid.value_counts())
    
    
```

### 2. StratifiedKFold 교차검증

```python
skf = StratifiedKFold(n_splits=3)
```

```python
count=0
for train_index, valid_index in  skf.split(iris_data,iris_target):# index 반환
    count+= 1
    label_train = df['species'].iloc[train_index]
    label_valid = df['species'].iloc[valid_index]
    #print(label_train, label_valid)
    
    print('교차검증 순서:{}'.format(count))
    print('학습 데이터 분포:',label_train.value_counts())
    print('검증 데이터 분포:',label_valid.value_counts())
```
