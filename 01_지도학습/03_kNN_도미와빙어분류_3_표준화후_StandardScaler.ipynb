### 5. 2차원 입력 데이터 만들기 - 훈련 데이터는 무조건  2차원

### 6. 레이블(label , target) 만들기
* 도미: 1
* 빙어: 0

```python
fish_target=[1]*35 + [0]*14
fish_target
```

### 7.데이터를 쉽게 이해하기 위해 df로 변경한다.

```python
df=pd.DataFrame(fish_data)
df.columns=['length','width']
df['target']=fish_target
df
```

###  8. 데이터 분리 train_test_split 함수 사용

```python
from sklearn.model_selection import train_test_split
```

```python
fish_data = np.array(fish_data)
fish_target = np.array(fish_target)
# train_test_split(입력데이터, 타겟)
#  stratify 는 불균형 데이터 경우에 균형있게 자동으로 분리해준다. stratify=fish_target
X_train, X_test, y_train, y_test=train_test_split(fish_data, fish_target, test_size=0.2, random_state=1,stratify=fish_target)
```

```python
# 총 49개중에 39개가 훈련 데이터
X_train.shape,  y_train.shape
```

```python
X_test.shape,  y_test.shape
```

```python
plt.scatter(X_train[:,0], X_train[:,1])#훈련 데이터 (길이, 무게 )
plt.scatter(X_test[:,0], X_test[:,1]) #테스트 데이터(길이, 무게)

plt.xlabel("length")
plt.ylabel("width")
plt.legend(["train set","test set"])
plt.show()
```

### 9. 표준화 

```python
# z = (x - 평균)/표준편차
# 표준편차는 일반적으로 주요 용도는 표준화할 때 많이 사용됨.
from sklearn.preprocessing  import StandardScaler
```

```python
# 반드시 fit(X_train)한 훈련데이터의 평균과 표준편차를 사용해서 테스트데이터 및 새로운 데이터로 표준화해야 된다.
sc = StandardScaler()
sc.fit(X_train)# fit한 훈련데이터로 표준화
train_scaled = sc.transform(X_train)
test_scaled = sc.transform(X_test)
new_value_scaled = sc.transform([[25,150]])
```

### 표준화 후 시각화

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(test_scaled[:,0], test_scaled[:,1])
plt.scatter(new_value_scaled[:,0], new_value_scaled[:,1])
plt.xlabel("length")
plt.ylabel("width")
plt.legend(["train_scaled set","test_scaled set", "new_value_scaled"])
plt.rc("font", family="Malgun Gothic")
plt.show()
```

### 8. 모델 생성 및 훈련

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(n_neighbors=5)
# kn.fit(fish_data, fish_target)
# kn.fit(X_train, y_train)
kn.fit(train_scaled, y_train)
KNeighborsClassifier()
```

```python
# lable의 종류
kn.classes_
```
