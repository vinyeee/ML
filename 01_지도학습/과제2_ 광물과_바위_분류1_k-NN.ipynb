### 1. 데이터 분석

```python
sonar_data=df.iloc[:,:-1]
sonar_target=df.iloc[:,-1]
sonar_data.shape,sonar_target.shape
```

```python
# heatmap
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
colormap = plt.cm.gist_heat   #그래프의 색상 구성.
plt.figure(figsize=(12,12))   #그래프의 크기.

# 그래프의 속성을 결정합니다. vmax의 값을 0.9로 지정해 0.9에 가까울 수록 밝은 색으로 표시.
sns.heatmap(df.corr(),linewidths=0.1,vmax=0.9, cmap=colormap, linecolor='white', annot=True)
plt.show()
```

### 2. 훈련 데이터 테스트 데이터 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(sonar_data,sonar_target,test_size=0.2)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

### 3. 모델 생성 및 훈련 - 비표준화

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(X_train, y_train)
```

```python
kn.classes_
```

### 4. 예측

```python
pred1=kn.predict(X_train)
pred1
```

```python
pred2=kn.predict(X_test)
pred2
```

### 5.평가

```python
kn.score(X_train,y_train)
```

```python
kn.score(X_test,y_test)
```

### 6. 표준화

```python
from sklearn.preprocessing  import StandardScaler
```

```python
sc= StandardScaler()
sc.fit(X_train)
train_scaled=sc.transform(X_train)
test_scaled=sc.transform(X_test)
```

### 7. 표준화 후 재훈련 및 예측, 평가

```python
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(train_scaled, y_train)
```

```python
pred1=kn.predict(train_scaled)
pred1
```

```python
pred2=kn.predict(test_scaled)
pred2
```

```python
kn.score(train_scaled,y_train)
```

```python
kn.score(test_scaled,y_test)
```
