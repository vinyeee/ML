```python
'''
    교차검증 (cross validation)
    
    1. KFold API 활용
    
    KFold(n_splits=5,shuffle=False,random_state=None) 
    gen = KFold.split(훈련데이터) ==> generater 반환
    X_train,X_valid=next(gen) ==> index 반환
    
    2. 검증 데이터와 이전 실습의 test 데이터 구별필수
    spec 에서는 검증데이터 대신에 test 데이터로 명시되어 있음.
    
    우리는 마지막에 쓰는 test 데이터만 test데이터라 부르기로 함
    
    3. 특징: 불균형 데이터 경우에는 사용안됨 ==> Stratified KFold
    
'''
```

```python
import pandas as pd
import numpy as np
```

### 1. 데이터 수집

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
```

```python
iris=load_iris()
iris
```

```python
iris_data=iris.data
iris_data
```

```python
iris.feature_names
```

```python
iris_target=iris.target
iris_target
```

```python
iris.target_names
```

### 데이터를 쉽게 이해하기 위해 df로 변경한다.

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=iris.target
df['species_name']=iris.target_names[iris.target]
df
```

### 모델 생성

```python
from sklearn.tree import DecisionTreeClassifier #결정트리
dtc=DecisionTreeClassifier()
```

### 2. 교차검증

```python
# n_splits ==> 총 데이터  150개를 5개의 fold로 구분, 각 폴드마다 30개씩 저장된다. 
# kfold=KFold(n_splits=5,shuffle=False,random_state=None) 기본
# shuffle = False ==> data 섞지 않음
# random_state ==> seed 고정
kfold=KFold(n_splits=5,shuffle=False,random_state=None) 
```

```python
from sklearn.metrics import accuracy_score
score_list=[]
for train_index, valid_index in  kfold.split(iris_data):# index 반환
    
    #index 이용해서 실제 훈련 데이터 및 검증 데이터 얻기
    X_train, X_valid = iris_data[train_index],iris_data[valid_index]
    y_train,y_valid=iris_target[train_index],iris_target[valid_index]
    
    #모델 이용해서 훈련
    dtc.fit(X_train,y_train)
    
    #예측 
    pred = dtc.predict(X_valid)
    
    #평가 
    score=accuracy_score(y_valid,pred)
    
    score_list.append(score)
    
    
score_list
```

```python
np.mean(score_list)
```
